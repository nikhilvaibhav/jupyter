```python
#SEARCHING
#BINARY SEARCH

arr = [10,20,30,40,50,60]
x =  90

def BinarySearch(arr,l,r,x):
    if (r>=l):
        
        mid = l+(r-1)//2
    
        if arr[mid] == x:
            return mid
    
        elif arr[mid]>x:
            return BinarySearch(arr,l,mid-1,x)
        else:
            return BinarySearch(arr,mid+1,r,x)
    else: 
        return -1
    
    
result = BinarySearch(arr,0,len(arr)-1,x)
if result !=-1:
    print("present at" , result)
    
else:
    print("NOT PRESENT")
```

```python
# BUBBLE SORT

def bubblesort(arr, n):
    for i in range(n):
        
        # using n-i in j loop, since i is already sorted.
        for j in range(n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                
arr = [16,11,9,34,5]
n = len(arr)
bubblesort(arr, n)
print("sorted element is: ")
for i in range(n):
    print(arr[i], end =" ")
```

```python
#SELECTION SORT

arr = [5,4,3,2,1]
n = len(arr)

def selectionsort(arr, n):
    
    for i in range(n):
        minind = i
        for j in range(i, n):
            if arr[j] < arr[minind]:
                minind = j
        arr[i], arr[minind] = arr[minind], arr[i]
            
selectionsort(arr, n)
print(arr)
```

```python
#INSERTION SORT

def insertion_sort(a):
    for i in range(1,len(a)):
        key = a[i]
        
        j = i-1
        while j>=0:
            if key<a[j]:
                a[j+1] = a[j]
                a[j] = key
                j-=1
            else:
                break
                
a = [7,4,2,1]
insertion_sort(a)
print(a)
            
    
```

```python
#STACK implementation using array

class Stack:
    def __init__(self):
        self.stack = []
        
    def push(self, data):
        self.stack.append(data)
        
    def display(self):
        for i in self.stack:
            print(i, end = " ")
        
    def pop(self):
        if self.is_empty():
            print("No element in the stack to be popped.")
        data = self.stack[-1]
        del self.stack[-1]
        print ("\n popped element is: ", data)
        self.display()
        
    def peek(self):
        print ("\n Topmost element is",self.stack[-1])
    
    def is_empty(self):
        if len(self.stack) == 0:
            print("stack is empty")
    
    def size(self):
        print(len(self.stack))
    
if __name__ == "__main__":
    stack = Stack()
    stack.push(3)
    stack.push(4)
    stack.push(7)
    stack.push(1)
    stack.push(8)
    stack.push(10)
    stack.push(12)
    stack.display()
    stack.pop()
    stack.peek()
    print("Size of the stack is: ")
    stack.size()
    stack.pop()
    print("\n Size is")
    stack.size()
    
            
        
```
